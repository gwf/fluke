```python
def step(root,num,sNum):
    return -((sNum**root)-num)/(root*(sNum**(root-1)))

#Err calculates the Error
def calError(sNum ,num , root):
    Err = ((sNum**root-num)**2)
    return Err

def calStartingNum(num, root):
    return num/root
```

```python
def printLine(stepNum, sNum, sNumN, Err, roundNum, tableSpace = 15):
    print(str("{:<"+str(tableSpace)+"} {:<"+str(tableSpace)+"} {:<"+str(tableSpace)+"} {:<"+str(tableSpace)+"}").format(stepNum ,round(sNum,roundNum) ,round(sNumN,roundNum) ,round(Err,roundNum)))
```

```python
def root(root,num,roundNum = 8):
    #roundNum is how many decmials are wanted
    #Intialize stepNum to zero (counts which step we are on)
    stepNum = 0
    sNum = calStartingNum(num, root)
    Err = calError(sNum, num, root)
    #Strating number ^ root
    sNumN = sNum**root
    
    #Table heeader line 
    print ("{:<15} {:<15} {:<15} {:<15}".format('Step Num','x','x^n','Err'))
    #First line of the table
    printLine(stepNum, sNum, sNumN, Err, roundNum, 15)
    
    #While the error to decimal amount isnt zero
    while(round(Err,roundNum) !=  0):
        #update the varaiables
        stepNum += 1
        sNum += step(root,num,sNum)
        sNumN = sNum**root
        Err = calError(sNum, num, root)
        #Print a table line
        printLine(stepNum, sNum, sNumN, Err, roundNum, 15)
        
    
```

```python
def multiRoot(low, up, roundNum = 6):
    counter = 0
    for up in range(low, up + 1):
        print("root(" + str(low) + ", " + str(up) + ")")
        root(low, up, roundNum)
```
